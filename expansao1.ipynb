```python

import cv2
import numpy as np
from matplotlib import pyplot as plt

#se carga la imagen en formato jpg por medio de la libreria cv2
img = cv2.imread('sources/foto.jpg')#Nota, el nombre de la imagen debe de coincidir con el mostrado en el codigo

plt.imshow(img)
plt.show()


#Se declara una lista con el espectro a revisar, en este caso es rgb
color = ('b','g','r')

#Se crea un ciclo for para realizar el histograma rgb en base a la imagen dada arriba
for i, c in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])

#Se muestra el histograma terminado
plt.show()


img_to_yuv = cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
img_to_yuv[:,:,0] = cv2.equalizeHist(img_to_yuv[:,:,0])
hist_equalization_result = cv2.cvtColor(img_to_yuv, cv2.COLOR_YUV2BGR)
 
plt.imshow(hist_equalization_result)
plt.show()

color = ('b','g','r')

#Se crea un ciclo for para realizar el histograma rgb en base a la imagen dada arriba
for i, c in enumerate(color):
    hist = cv2.calcHist([hist_equalization_result], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])

#Se muestra el histograma terminado
plt.show()

cv2.destroyAllWindows()


```
